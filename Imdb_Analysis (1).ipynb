```python
import numpy as np    
import pandas as pd
import sqlite3
```

First, import the sqlite3 module and then create a connection object which will connect us to the database and will let us execute the SQL statements.

```python
db='movies.sqlite'
conn=sqlite3.connect(db)
cur=conn.cursor()        #cursor() - cursor object is used to call the execute() method to execute any SQL queries.
```

# Ready to go

Question1:Can you get all data about movies

```python
cur.execute('SELECT * FROM movies')
```

```python
movies = cur.fetchall()  #fetchall()-fetch all rows for the current query, if no rows avaliable then,it returns an empty list.
```

```python
movies = pd.DataFrame(movies, columns = ['id', 'original_title', 'budget', 'popularity', 'release_date', 'revenue', 'title', 'vote_average', 'vote_count', 'overview', 'tagline', 'uid', 'director_id'])
print('Shape of Movies data :',movies.shape)
print('\n\n\n')
movies.head(10)
```

```python
movies.info()
```

```python
import matplotlib.pyplot as plt #for graph designing.
import seaborn as sns
%matplotlib inline 

conn = sqlite3.connect('movies.sqlite')
query = "SELECT * FROM movies;"
db = pd.read_sql_query(query, conn)
conn.close()

db.head()
```

```python
# Simple line plot

x=db['id']
w=0.3
bar1=np.arange(len(x))
bar2=[i+w for i in bar1]
budget=db['budget']
revenue=db['revenue']
plt.xlabel('Title')
plt.ylabel('movies budget and revenue')
plt.bar(bar1,budget,w,label="budget")
plt.bar(bar2,revenue,w,label="revenue")
plt.xticks(bar1,x)

plt.ylim(0,900000000)

plt.show()
```
